```python
import pandas as pd
import googlemaps
```

```python
df = pd.read_excel(r"C:\Users\jonat\Desktop\ShopAddresses.xlsx") #past file path, exp:"C:\Users\jonat\Desktop\addresses.xlsx"
df
```

```python
gmaps_key = googlemaps.Client(key = "") # paste api key in double quotes
```

```python
df.iat[1,1] #ensure correct loop correct row
```

```python
for i in range(0, len(df), 1): #Ensure loop trhough correct row
    print(df.iat[i,1])
```

```python
df["LAT"] = None
df["LON"] = None

for i in range(0, len(df), 1):
    geocode_result = gmaps_key.geocode(df.iat[i,1])
    try:
        lat = geocode_result[0]["geometry"]["location"]["lat"]
        lon = geocode_result[0]["geometry"]["location"]["lng"]
        df.iat[i, df.columns.get_loc("LAT")] = lat
        df.iat[i, df.columns.get_loc("LON")] = lon
    except:
        lat = None
        lon = None
```
